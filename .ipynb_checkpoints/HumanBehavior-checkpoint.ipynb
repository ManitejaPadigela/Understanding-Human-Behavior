```python
def smoke(cluster):
    if cluster=='Yes':
        return 1
    else:
        return 0
```

```python
human_data['Smoke'] = human_data['Do you Smoke'].apply(smoke)
```

```python
def dexterity(cluster):
    if cluster=='Right Hand':
        return 1
    elif cluster=='Left Hand':
        return 0
    else:
        return 2
```

```python
def drink(cluster):
    if cluster=='Yes':
        return 1
    else:
        return 0
```

```python
human_data['Hand Preffered'] = human_data['Dexterity'].apply(dexterity)
human_data['Alcohol Intake'] = human_data['Do you Drink'].apply(drink)
```

```python
human_data.head()
```

```python
human_data.drop("Timestamp", axis=1, inplace=True)
human_data.drop("Do you Drink", axis=1, inplace =True)
human_data.drop("Dexterity", axis=1, inplace=True)
```

```python
human_data.head()
```

```python
def age(cluster):
    if cluster=='<18':
        return 0
    elif cluster=='>18 and <30':
        return 1
    else:
        return 2
    
def bedtime(cluster):
    if cluster=='Before 10:30PM':
        return 1
    else:
        return 0
    
def exercise(cluster):
    if cluster=='Yes':
        return 1
    else:
        return 0
    
def drinktype(cluster):
    if cluster=='Tea or Coffee':
        return 0
    elif cluster=='Milk':
        return 1
    elif cluster=='Fruit juices':
        return 2
    
def typeSpent(cluster):
    if cluster=='Movies':
        return 0
    elif cluster=='Series':
        return 1
    elif cluster=='Reading Books':
        return 2
    
def typeGame(cluster):
    if cluster=='Physical':
        return 0
    elif cluster=='Mobile or PC':
        return 1
    elif cluster=='Both':
        return 2
    
def partying(cluster):
    if cluster=='No':
        return 0
    elif cluster=='Yes':
        return 1
    
def food(cluster):
    if cluster=='Vegetarian (if egg)':
        return 0
    elif cluster=='Non- Vegetarian':
        return 1

def maritalStatus(cluster):
    if cluster=='Single':
        return 0
    elif cluster=='Married':
        return 1


def empStatus(cluster):
    if cluster=='Employed or Self Employed':
        return 0
    elif cluster=='UnEmployed':
        return 1
    elif cluster=='Student':
        return 2
    
def stay(cluster):
    if cluster=='Hostel':
        return 0
    elif cluster=='Home':
        return 1

def timeType(cluster):
    if cluster=='Night Owl':
        return 0
    elif cluster=='Early Bird':
        return 1
```

```python
human_data['Aged'] = human_data['Age'].apply(age)
human_data['Before 10:30PM'] = human_data['Bed time'].apply(bedtime)
human_data['Exercise'] = human_data['Do you Exercise'].apply(exercise)
human_data['Drink type'] = human_data['Most Preferred'].apply(drinktype)
human_data['Entertainment'] = human_data['Mostly time Spent on'].apply(typeSpent)
human_data['Games type'] = human_data['Which games do you play or Preferred'].apply(typeGame)
#human_data['Outdoor Partying'] = human_data['Regular Partying or Social Gathering'].apply(partying)
human_data['Food Type'] = human_data['Food Preferred'].apply(food)
human_data['Marriage status'] = human_data['Marital Status'].apply(maritalStatus)
human_data['Employment type'] = human_data['Employment Status'].apply(empStatus)
human_data['Stays'] = human_data['Stays at'].apply(stay)
human_data['Time Type'] = human_data['Are you'].apply(timeType)
```

```python
human_data.head()
```

```python
human_data.drop("Age", axis=1, inplace=True)
human_data.drop("Bed time", axis=1, inplace =True)
human_data.drop("Do you Exercise", axis=1, inplace=True)
human_data.drop("Most Preferred", axis=1, inplace=True)
human_data.drop("Mostly time Spent on", axis=1, inplace =True)
human_data.drop("Which games do you play or Preferred", axis=1, inplace=True)
human_data.drop("Food Preferred", axis=1, inplace=True)
human_data.drop("Marital Status", axis=1, inplace =True)
human_data.drop("Employment Status", axis=1, inplace=True)
human_data.drop("Stays at", axis=1, inplace =True)
human_data.drop("Are you", axis=1, inplace=True)
#human_data.drop("Regular Partying or Social Gathering", axis=1, inplace=True)
```

```python
human_data.head()
```

```python
final_human_data = human_data[['Male','Before 6:30 am','Before 10:30PM','Smoke','Exercise','Entertainment','Hand Preffered'
                              ,'Alcohol Intake','Aged','Drink type','Games type','Food Type','Marriage status','Employment type',
                              'Stays','Time Type']]
```

```python
final_human_data.head()
```

```python
y = final_human_data['Before 10:30PM']
```

```python
X = final_human_data[['Before 6:30 am','Entertainment','Food Type','Stays','Employment type','Exercise']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
print('Coefficients: \n', lm.coef_)
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
coeffecients = pd.DataFrame(lm.coef_,X.columns)
coeffecients.columns = ['Coeffecient']
coeffecients
```
